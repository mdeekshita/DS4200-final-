```python
import pandas as pd
import altair as alt

# Load your dataset
data = pd.read_csv("updated_dataset.csv")

# Standardize the column for filtering
data["Commendation or Complaint"] = data["Commendation or Complaint"].str.lower()

# Replace quarter numbers with season names
data["Season"] = data["Quarter"].replace({
    1: "Winter", 2: "Spring", 3: "Summer", 4: "Fall"
})

# Aggregate the data
agg_data = data.groupby(["Season", "Commendation or Complaint"]).size().reset_index(name="count")

# Create a selection dropdown
selection = alt.selection_single(
    fields=["Commendation or Complaint"],
    name="Report Type",
    bind=alt.binding_select(options=agg_data["Commendation or Complaint"].unique())
)

# Create the chart
chart = alt.Chart(agg_data).mark_bar().encode(
    x=alt.X("Season:N", title="Season", sort=["Winter", "Spring", "Summer", "Fall"]),
    y=alt.Y("count:Q", title="Number of Reports"),
    color=alt.Color("Season:N", legend=None),
    tooltip=[
        alt.Tooltip("Season:N", title="Season"),
        alt.Tooltip("count:Q", title="Number of Reports")
    ]
).add_selection(
    selection
).transform_filter(
    selection
).properties(
    title="NYC Transit System Reports by Season",
    width=400,
    height=300
)

chart.save("nyc_transit_chart.html")
chart
```
